```python
%time df2 = pd.read_csv('pickles_and_munged_csv/final_data_v3.csv', engine='c', usecols=[1,2,3,4], dtype={'publn_nr':str,'appln_id':str})
# don't need publn
# pickling is so so slow...
```

```python
# get list of unique codes and sort them
unique_codes = df2.ipc_class_symbol.unique()
```

```python
unique_codes.sort()
unique_codes
```

```python
code_ints = range(len(unique_codes))
zipped = zip(unique_codes,code_ints)
zipped
```

```python
code_dict = {code: code_int for code, code_int in zipped}
code_dict['A47J  44']
```

```python
pk.dump(code_dict, open('pickles_and_munged_csv/code_dict_v1.pk','wb'))
```

```python
def code_to_int(code):
    return code_dict[code]
```

```python
df2.ipc_class_symbol.map(code_to_int)
```

```python
df2.ipc_class_symbol = df2.ipc_class_symbol.map(code_to_int)
```
